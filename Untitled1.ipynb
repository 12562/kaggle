```python
import io
import pandas as pd
dftrain = pd.read_csv(io.BytesIO(uploaded['train.csv']))
dfeval = pd.read_csv(io.BytesIO(uploaded['test.csv']))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.compat.v2.feature_column as fc
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
print(dftrain)
print(len(dftrain['Ticket'].unique()))
print(dftrain['Ticket'].isnull().any())

print("Len Embarked unique",len(dftrain['Embarked'].unique()))
print("Num embarked null entries",sum(dftrain['Embarked'].isnull()))

print("Len Cabin unique:",len(dftrain['Cabin'].unique()))
print("Num Cabin null entries",sum(dftrain['Cabin'].isnull()))

print ("Len Pclass unique:", len(dftrain['Pclass'].unique()))
print (dftrain['Pclass'])

```

```python
y_train = dftrain.pop('Survived')
passenger_id = dftrain.pop('PassengerId')
name = dftrain.pop('Name')
#ticket = dftrain.pop('Ticket')
fare = dftrain.pop('Fare')
#cabin = dftrain.pop('Cabin')
#ticket_eval = dfeval.pop('Ticket')
fare_eval = dfeval.pop('Fare')
#cabin_eval = dfeval.pop('Cabin')

mean_age = dftrain['Age'].mean(axis=0)
std_age = dftrain['Age'].std(axis=0)
#mean_fare = dftrain['Fare'].mean(axis=0)
#std_fare = dftrain['Fare'].std(axis=0)
print(mean_age)
```

```python
dftrain['Age'] = dftrain['Age'].fillna(mean_age)
#dftrain['Fare'] = dftrain['Fare'].fillna(mean_fare)
dftrain['Age'] = (dftrain['Age'] - mean_age) / std_age
#dftrain['Fare'] = (dftrain['Fare'] - mean_fare) / std_fare
dftrain['Cabin'] = dftrain['Cabin'].fillna('U')
dftrain['Embarked'] = dftrain['Embarked'].fillna('U')
dftrain['Ticket'] = dftrain['Ticket'].fillna('U')

dfeval['Age'] = dfeval['Age'].fillna(mean_age)
#dfeval['Fare'] = dfeval['Fare'].fillna(mean_fare)
dfeval['Age'] = (dfeval['Age'] - mean_age) / std_age
#dfeval['Fare'] = (dfeval['Fare'] - mean_fare) / std_fare
dfeval['Cabin'] = dfeval['Cabin'].fillna('U')
dfeval['Embarked'] = dfeval['Embarked'].fillna('U')
dfeval['Ticket'] = dfeval['Ticket'].fillna('U')
```

```python
print(dftrain['Ticket'].unique())

import re
def find_nonnum(text):
    #print(text)
    result = re.findall("[^0-9]",text)
    result = "".join(result)
    if ( result == ''):
      result = "N"
    #print(result)
    return result
dftrain['Ticket']=dftrain['Ticket'].apply(lambda x: find_nonnum(x))
print(dftrain['Ticket'].unique())
```

```python
X_train, X_validation, Y_train, Y_validation = train_test_split(dftrain, y_train, test_size=0.20, random_state=1, shuffle=True)

CATEGORICAL_COLUMNS = ['Sex', 'Cabin', 'Embarked',  'Ticket', 'Pclass', 'SibSp', 'Parch' ]
#CATEGORICAL_COLUMNS = ['Sex', 'Ticket', 'Embarked', 'Pclass', 'SibSp', 'Parch']
#NUMERIC_COLUMNS = ['Fare', 'Age']
NUMERIC_COLUMNS = ['Age']

print(X_train['Embarked'].unique())
print(Y_train.shape)

feature_columns = []
for feature_name in CATEGORICAL_COLUMNS:
    vocabulary = dftrain[feature_name].unique()
    feature_columns.append(tf.feature_column.categorical_column_with_vocabulary_list(feature_name, vocabulary))

for feature_name in NUMERIC_COLUMNS:
    feature_columns.append(tf.feature_column.numeric_column(feature_name, dtype=tf.float32))
                                                                                             
```

```python
def make_input_fn(data_df, label_df, num_epochs=10, shuffle=True, batch_size=32):
    def input_function():
        ds = tf.data.Dataset.from_tensor_slices((dict(data_df), label_df))
        if shuffle:
           ds = ds.shuffle(1000)
        ds = ds.batch(batch_size).repeat(num_epochs)
        return ds
    return input_function


def make_predict_fn(data_df):
    def input_function(batch_size = 32):
        ds = tf.data.Dataset.from_tensor_slices((dict(data_df), None))
        ds = ds.batch(batch_size)
        return ds
    return input_function

train_input_fn = make_input_fn(X_train, Y_train)
validation_input_fn = make_input_fn(X_validation, Y_validation, num_epochs=1, shuffle=False)
eval_input_fn = make_predict_fn(dfeval)


linear_est = tf.estimator.LinearClassifier(feature_columns=feature_columns)

linear_est.train(train_input_fn)

result = linear_est.evaluate(validation_input_fn)

print(result['accuracy'])
result = list(linear_est.predict(validation_input_fn))
actual_result = list(linear_est.predict(eval_input_fn))

```

```python
#print(actual_result[0:1]['probabilities'][1])

res = [int(i['probabilities'][1] > .5) for i in actual_result]
print(res)
pid = list(dfeval['PassengerId'])
df = pd.DataFrame(list(zip(pid,res)),columns=['PassengerID', 'Survived'])
df.to_csv('file', index=False)
!head file
```
